Contour plot is used to present meteorological or geographic data. It is also used to present density, brightness, and electrical potential and widely used in data analysis and machine learning. It shows how a response variable relates to two predictor variables.

Contour Plots gives a 2D view where all the points with the same answer are connected by a line. This line is a contour line. The main elements of a contour plot:
* x-axis and y-axis show predictors
* Contour lines to represent the same response values
* Coloured bands that provide a range of response values.

```python
import numpy as np
import matplotlib.pyplot as plt
import pylab
```

Now we will generate some values to generate the plot:

```python
xp = np.arange(-8, 10, 2)
yp = np.arange(-8, 10, 2)
zp = np.ndarray((9,9))
```

Now we will replace the random values that we can come across while generating values:

```python
for x in range(0, len(xp)):
    for y in range(0, len(yp)):
        zp[x][y] = xp[x]**2 + yp[y]**2
```

So now we have the data ready letâ€™s start by plotting a very simple contour plot:

```python
plt.figure(figsize=(7, 5))
plt.title('Contour Plot')
contours = plt.contour(xp, yp, zp)
plt.clabel(contours, inline=1, fontsize=12)
plt.show()
```

### Plotting A Saddle Surface
Weâ€™ll do the same procedure with a slightly different formula for the z values:

```python
xp = np.arange(-3, 4)
yp = np.arange(-3, 4)
zp  =np.ndarray((7,7))
for x in range(0, len(xp)):
    for y in range(0, len(yp)):
        zp[x][y] = xp[x]*xp[x] - yp[y]*yp[y]
```

```python
plt.figure(figsize=(8, 6))
plt.title('Contour plot for saddle surface - Hyperbolic paraboloid')
contours = plt.contour(xp, yp, zp)
plt.clabel(contours, inline=1, fontsize=12)
plt.show()
```

### Plotting Density Contour Plots

Now weâ€™re going to create a different dataset where the data points are closer to plot a density graph in contour plots. Here are the values of our data:

```python
x = np.linspace(0, 5, 60)
y = np.linspace(0, 5, 48)
```

Now we have created values for `x` and `y` in the above code, we will now create values for `z` by creating a function:

```python
def fn(x, y):
    return np.sin(x)**5 + np.cos(y+17)**8
```

```python
X, Y = np.meshgrid(x, y)
Z = fn(X, Y)
```

The data is now ready. Now letâ€™s draw the density ploy and add a colour palette also specifying that we want 15 equally spaced lines drawn in the range:

```python
plt.figure(figsize=(8, 6))
contours = plt.contour(X, Y, Z, 15, cmap = 'RdGy')
plt.clabel(contours, inline=True, fontsize=12)
plt.show()
```

There is a function in matplotlib known as contourf. Letâ€™s see what we can do by using this function:

```python
plt.figure(figsize=(14, 10))
plt.contourf(X, Y, Z, 20, cmap = 'RdGy')
plt.colorbar()
plt.show()
```
