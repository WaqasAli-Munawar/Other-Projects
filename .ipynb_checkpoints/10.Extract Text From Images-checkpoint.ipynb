In this file, we will use the power of Python to extract text from images. The applications of this technique are endless. Some examples include:
* Data mining for machine learning tasks

To extract text from images we’ll be using a library called Python **Tesseract**. **Tesseract** is an **optical character recognition** tool for python. In simple words, by using this package we can **recognize and read** the text embedded in the images.

### Setting Up Tesseract

* Install the Python library
* Install the Tesseract app
